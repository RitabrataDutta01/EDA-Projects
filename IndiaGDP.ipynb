```python
%%markdown
Exploratory Data Analysis on GDP Per Capita | Gov Expenditure | Trade dataset

We are going to be plotting GDP per Capita v/s Year graph on India.

First we have to import the necessary librarires and load our dataset into a pandas DataFrame
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('gdp.csv')
```

```python
%%markdown
Here is the dataset which we are going to be working with,
```

```python
df
```

```python
%%markdown

Next Step is that we are going to clean the data set as to remove any outliers or missing values and separating the part 
which contains the information about India into a separate DataFrame.
```

```python
df =df.dropna()
idf = df[df['Entity'] == 'India']
idf.head()
```

```python
%%markdown

Now that we have separated the information aboutIndia into a separate table, we are going to be plotting the yearwise GDP per capita 
and find its growth trends, be it increasing or decreasing.
```

```python
plt.figure(figsize = (20,30))
plt.plot(idf['Year'], idf['GDP per capita'])
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('Year v/s GDP per Capita from 1960 to 2022 for India')
plt.grid(True)

years = idf['Year'].unique()
if len(years) > 15:
    plt.xticks(years[::5], rotation=45, ha='right')
else:
    plt.xticks(years, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
%%markdown

Conclusion-

From the graph it is evident that even though there are some small downwards spikes in between 1960-2000, the gdp growth rate of India
is exponential and appears to particularly accelerate in between 1985 and 1990. There were mostly upwards spikes indicating a sharp GDP grwpth and
most downward spikes spikes were during the itmes of conflict with Pakistan, but there are no signs of stagnation of India's GDP growth.

Further analysis could open up new doors for India to futher increase its GDP growth rate and surpass other countries whcih are higher on the 
GDP list.
```
