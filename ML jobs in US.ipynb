Introduction:-

Job Market nowadays is one of the most competetive sectors after graduation. Machine Learning is a branch of Computer Science and has a lot of job market but an even higher
amount of candidates. In this EDA report we are going to be analyzing the ML job market in USA using the kaggle dataset MACHINE LEARNING JOB POSTINGS IN THE US.

First step is to import all the necessary libraries and to load the data to be in a usable format.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./data/1000_ml_jobs_us.csv')

df.head()
```

From the given dataset we can see that it has 9 features, from here we can make multiple plots to find whic posts are mostly offered and also where the most jobs are geting concentrated.

Next Step is to clean the data by dorpping features which are not so useful to reduce the amount of features which we have to analyze.

```python
df = df.drop(columns=['company_address_locality', 'company_website', 'company_website', 'job_description_text'])
df
```

Now that we have reduced the number of features, we can now proceed with the first plot to see from where the most amount of jobs were posted to get a rough idea of how the job market is distributed

```python
data = df['company_name']

company = data.value_counts()

company
```

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title('Top 10 Companies Hiring Machine Learning Professionals')
plt.xlabel('Company Name')
plt.ylabel('Number of Jobs')
sns.barplot(x=company.index[:10], y=company.values[:10], palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

From the above graph we can conclude that most ML engineers are being hired by companies which provide some cloud based service which require to try to cater to an individuals personal interests like entertainment media like short videos/ movies or food etc. Although there are some anomalies like Splunk and Adobe which are now focusing more on AI and ML as thier USP.

Similarly we can plot graphs to see the trends which affect the job positions offered and in what regions most of the job market is concentrated.

```python
positions = df['job_title']
positions = positions.value_counts()
positions
```

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title('Top 10 Job Titles for Machine Learning Professionals')
plt.xlabel('Job Title')
plt.ylabel('Number of Jobs')
sns.barplot(x=positions.index[:10], y=positions.values[:10], palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

From the above plot we can see what positions are being offered the most by the companies. At the top with the most job offers we have Machine Learining Engineer, which actually comprises around 25% of the total jobs posted.

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title('Titles not so much in demand')
plt.xlabel('Job Title')
plt.ylabel('Number of Jobs')
sns.barplot(x=positions.index[-10:], y=positions.values[-10:], palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

From the above plot we can see what positions are being offered the most by the companies the least amount. These posts will have high competetion to  get into due to such scarcity.

```python
location = df['company_address_region'].value_counts()
location
```

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title('Locations hiring Machine Learning Professionals')
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
sns.barplot(x=location.index[:10], y=location.values[:10], palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

From this plot it is evident that most ML jobs are from the California region with Central America being in Close second. This helps us to get a rough estimate as to where most of the job market is concentrated and will help applicants pin point the market dense areas.

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title("Regions not so much in demand")
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
sns.barplot(x=location.index[-10:], y=location.values[-10:], palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

The above plot shows us the regions from where the list number of Job Postings came from, shows that the job market here is extremely competetive with long term contracts

```python
experience = df['seniority_level'].value_counts()
experience
```

```python
plt.figure(figsize=(20, 10))
sns.set_style('darkgrid')
plt.title('Experience Levels in demand')
plt.xlabel('Experience Level')
plt.ylabel('Number of Jobs')
sns.barplot(x=experience.index, y=experience.values, palette='viridis')
plt.xticks(rotation=45)
plt.show()
```

From the above plot it is evident that Mid-Senior Level experience is the most in-demand for these jobs with Entry Level jobs coming up to second most demanded. We can also see that Director level experience is the least in demand with no offers for executive.
